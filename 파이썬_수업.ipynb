# python keywords rules
- 파이썬에서 예약된 단어_구문(syntax)와 구조(structure)를 정의하는데 사용
1. 소문자, 대문자, 숫자,_(underscore)만 가능(대소문자 구분!)
2. 숫자로 시작 안된다
3. 키워드는 식별자로 사용 불가하다
4. 특정 심벌 사용 불가하다(ex)!, @, #, $)
5. 길이는 상관없다



# python statement
        ( 원칙 : 한 줄에 한 명령어씩 작성한다.)

## Multi-line statement

- 명령어 길이가 길어 여러줄로 작성해야 할 경우

    - 괄호로 묶는다/ 역슬래시'\'을 사용한다

- 명령어 길이가 짧아 한줄로 작성할 경우

    - 세미콜론';'으로 구분한다.


## Indentation(들여쓰기)
- 코드의 블록을 정의함(하나로 묶어 진행)
- '1->2->if-> 3' 여기서 IF 를 만족해야 3을 진행할 수 있다




## Comments(주석)
- #(hash)으로 구성
- ''' 또는 """기호로 다중 주석 처리
- 개발자용 메모

## Docstrings(documentation strings) in python
- 나중에 필요할 때 꺼내 보려고 코드 안에 저장해둔 글
- ex) """Function to double the value""" 으로 저장
- __doc__속성(attribute)으로 객체와 연결됨


## Variables(변수)
- 메모리에 데이터를 저장하기 위해 사용되는 지정된 위치

## Constants(상수)
- 파이썬에서는 상수가 존재하지 않음
- 상수는 대문자,_ 기호의 조합으로 작성
 -  ex) PI = 3.14  ????

# Literal
#numeric(숫자형): int(정수), float(실수), complex(복소수,j)
#string(문자열): ' ' 나 " " 사용
#boolean:True(참,1), False(거짓,0)(첫 글자는 대문자로 약속)

##list
- [  ,  ]대괄호 사용/ 순서지정/ item 변경 가능/index는 0부터 시작/
##tuple
- (  ,  )소괄호 사용/ 순서 지정/item 변경 불가능 index는 0부터 시작
##set
- {  ,  }중괄호 사용/ 정렬되지 않은 모음/순서 없음/아이템_add(), update(), discard(), remove()를 통해 추가 제거 가능/slicing operation 통해 접근 불
##dictionary
- {key: value1} 중괄호 사용/pair로 정렬되지 않은 모음/key를 통해 value 검색 가능/ 순서 없엄

##string
- 유니코드 문자들의 나열
- list, tuple 처럼 slicing연산자 [] 를 사용하여 접근 가능
- slicing operation 통해 변경 불가
- enumerate()을 통해 인덱스와 값의 쌍 생성 가능

##enumerate()
- 어느 위치에 무엇이 있는지 확인
- index와 해당하는 값을 알려줌

# Type casting(형변환)_암묵적/명시적

## Implicit(암시/암묵적 변경)
- 자동으로 형 변환 실시
- 사용자 개입이 필요 없음
- ex) 3 + 2.0 = 5.0
## Explicit(명시적 변경): int(), float(), str() 등
- 개발자가 의도를 가지고 변환 실시

## print()
print(*objects, sep=' ', end ='\n')
-> * 임의의 정보를 전달 받을 수 있게 저장
-> 사이사이에 '  ' 을 넣어 출력
-> 끝나면 '\n' 수행


##str.format()
- print('A'.format B)
-> '.'점을 찍어 A를 프린트 하는데 B 기능을 쓸 수 있게 됨.
- A 안에 {} 을 사용하여 B 가 작동하도록 함

--> lower(), upper(), join(), split(), fing(), replace() 예시 만들어 보자

## input()
- 무조건 문자열(string)로 저장됨
- 사용자로부터 data 입력 받음
- print() 와 반대 개념

#산술(arithmetic)연산자--> 결과는 숫자
'+'  add

'-'  subtract

'*'   multiply

'/'   divide

'%'    modulus-remainder of x/y(x%y)

'//'  floor division-나눗셈의 몫만 가져옴 print(x//y)

**   exponent- x to the power y (x**y)


#비교(comparison)연산자--> 결과는 논리값

'<'    less than

'=='    equal to

'!='     not equal to

'<='    less than or equal to
