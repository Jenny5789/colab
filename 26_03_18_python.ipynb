**📌 항등(identity)연산자** : 두 개의 값이 '동일한 주소'를 갖는지 확인하는데 사용

````python
x1 = 5            # 리터널, x1, y1 동일한 주소를 가짐
y1 = 5
x2 = 'Hello'
y2 = 'Hello'
x3 = [1,2,3]      # 리터널 모음, x3, y3 다른 주소를 가짐
y3 = [1,2,3]


# Output: False
print(x1 is not y1)  # x1의 주소는 y1의 주소가 아니다 ->(리터널이므로 동일한 주소를 갖는다) -> False

# Output: True
print(x2 is y2)      # x2의 주소는 y2의 주소다->(리터널이므로 동일한 주소를 갖는다) -> True

# Output: False
print(x3 is y3)      # x3의 주소는 y3의 주소다 -> (리터널의 모음이므로 다른 주소를 갖는다) -> False
````

````python
a = 10
b = [10,11]
print(a is b)    # a의 주소는 b다. -> a는 리터널, b는 리터널 모음--> 주소 다름 -> False
print(b[0])      # b의 item 중 '10'은 a의 '10'과 동일한 주소를 갖는다
print(a is b[0]) # a의 주소는 b의 10주소 이다-> a의 리터널과 b의 리터널 주소로 보면 다르지만
                 # b의 10을 아이템으로 보면 10 값이 동일하여 -> True
````

````python
def outer_function():
    b = 20                       # b : local(지역변수)
    def inner_func():
        c = 30                   # c : nested local
        print(a)
    inner_func()
a = 10                           # a : global(전역변수)
outer_function()
````

````python
def outer_function():
    a = 20                      # 지역변수
    def inner_function():
        a = 30                  # 지역변수
        print('a =', a)
    inner_function()
    print('a =', a)
a = 10                        # 전역변수는 지역변수에서도 사용 가능함
                              # 파이썬에서는 전역 변수명을 바꿔 지역변수에서 사용 가능. global을 이용
outer_function()
print('a =', a)
````

````python
def outer_function():
    global a                   # a -> 전역 변수값(10) 가져옴
    print('a1 =', a)
    a = 20                     # a = 20
    print('a2 =', a)
    def inner_function():
        global a               # a -> 전역 변수값(20) 가져옴
        print('a3 =', a)
        a = 30                 # a = 30
        print('a4 =', a)
    inner_function()
    print('a5 =', a)
a = 10
print('a6 =', a)
outer_function()
print('a7 =', a)
````

🔍 global - 전역 변수에 있는 값을 새롭게 변수로 취급

🔍 nonlocal - 직전(namespace) 지역 변수값을 적용 시킴

````python
def outer_function():
    a = 20                  # a = 20
    print('a1 =', a)
    def inner_function():
        nonlocal a          # a -> 지역 변수값(20) 가져옴
        print('a3 =', a)
        a = 30              # a = 30
        print('a4 =', a)
    inner_function()
    print('a5 =', a)        # a = 30
a = 10
print('a6 =', a)            # a = 10
outer_function()
print('a7 =', a)            # 전역 변수 a = 10 은 함수 끝날 때까지 고정
````

````python
def outer_function():
    global a                # a = 10
    a = 20                  # a = 20
    def inner_function():
        global b
        b = 30              # global-> b = 30으로 전역변수 설정
    inner_function()
    print('b =', b)         # b = 30 print

a = 10
outer_function()
print('a =', a)             # a = 20
````

# 📝 파이썬 흐름 제어(flow control)

**📝 조건문**: if

**📌 if** : 조건 만족하는 경우에**만** 실행

```
# if test expression:
    statements(s)
```



````python
num = 3
if num > 0:
    print(num, "is a positive number.")    # 조건 num > 0 만족하여 print
print("This is always printed.")
num =-1                                   # namespace가 맨 앞(없을경우) = 새로운 조건 시작
if num > 0:
    print(num, "is a positive number.")   # 조건 num > 0 만족 못 하여 print 안됨
print("This is also always printed.")     # 맨앞 namespace = 조건 시작->  print
````

**📌 if....else** : 조건 만족하는 경우와 아닌 경우를 구분하여 실행 내용이 달라짐


```
# if test expression:
    Body of if
else:
    Body of else
```



````python
num = 3

if num >= 0:
    print("Positive or Zero")    # num >= 0 만족하여 print
else:
    print("Negative number")
````

**📌 if...elif....else** : 여러 개의 조건 만족 여부를 판단하고, 각 조건에 따라 실행 내용이 달라짐

```
# if test expression:
    Body of if
elif test expression:
    Body of elif
else:
    Body of else
```



````python
num = 3.4

if num > 0:
    print("Positive number")         # num = 3.4로 > 0 만족 --> print
elif num == 0:
    print("Zero")
else:
    print("Negative number")
````

````python
num = 0

if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")                    # num = 0로 = 0 만족 --> print
else:
    print("Negative number")
````

````python
num = -3.4

if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")    # num = -3.4로 < 0 만족 --> print
````

````python
num = float(input("Enter a number: "))     #input -> 문자열로 가져옴 / float()--> 실수형으로 num에 대입시킴

if num >= 0:                               # 0 보다 크면 아래 조건 실행
    if num == 0:                            # 0 이면 zero print
        print("Zero")
    else:                                   # 0 이 아니면 positive number print
        print("Positive number")
else:
    print("Negative number")              # 0 보다 크지 않으면 negative number print
````

📌 while loop, while loop + else
- while 루프는 조건을 만족한다면 계속 실행

```
# while text_expression:
       Body of while
```



````python
n = int(input("Enter n: "))
sum = 0                       # sum은 0으로 초기화
i = 1                         # i = 1
while i <= n:                 # i가 n 보다 작을 때
    sum = sum + i             # sum = 0 + 1 ....
    i = i+1                   # i에 1씩 더해 무한 루프 돌지않게 조건을 만든다.
print("The sum is", sum)
````

````python
counter = 0                     # counter = 0 으로 초기화

while counter < 3:              # counter가 0보다 작으면 아래 실행
    print("Inside loop")        # print
    counter = counter + 1       # 무한 루프 방지 조건
else:                           # counter가 0보다 작지 않으면 실행
    print("Inside else")        # print
````

📌 break : 반복 루프를 종료 시킴/THE END

📌 continue : 반복 루프를 skip 시킴/ SKIP AND CONTINUE

````python
for val in "banana":
    if val == "a":
        print(val)         # 'a'찾음
        break              # break -> 중단시킴     # 'break'는 조건문 밑에 위치해야 한다.
print("The end")
````

````python
for val in "banana":
    if val == "a":
        print(val)       # 'a' 찾음
        continue         # continue-> 이번 skip하고 다시 시작     # 'continue'는 조건문 밑에 위치해야 한다.
print("The end")
````

# 📝 함수(function)

📌 function : 특정 작업을 수행하기 위한 관련된 명령문의
- 선언 양식

```
# def function_name(parameters):
    “““ docsting ”””
    statement(s)
```
- 사용하기 이전에 함수가 선언되어 있어야 한다




````python
def greet(name):                               # greet(name) 함수로 정의
    """This function greets to
    the person passed in as
    a parameter"""
    print("Hello, " + name + ". Good morning!")    # 사용하기 전에 선언됨
greet('Paul')
````

- return 명령문 사용 가능함

```
# return [expression_list]
```

````python
def absolute_value(num):                  # absolute_value 함수로 정의
    """This function returns the absolute
    value of the entered number"""
    if num >= 0:
        return num
    else:
        return-num
num = absolute_value(2)
print(num)                              # print(absolute_value(2))--> 2 print
print(absolute_value(-4))               # print(absolute_value(4))--> 4 print
````

````python
- 변수의 적용범위는 함수 내부와 외부가 별개이다.
````

````python
def my_func():
    x = 10
    print("Value inside function:",x)     # my_func() 내부 함수값인 x = 10 이 입력된다.
x = 20
my_func()
print("Value outside function:",x)        # my_func() 외부 함수값인 x = 20 이 입력된다.
````
