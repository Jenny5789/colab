## 🔹vi 사용법_내용 수정
| 명령키(명령어)|  | 기능|
| :---: | :---: | :--- |
|r||커서가 위치한 글자를 다른 글자로 수정|

|s, #s||커서 위치부터 esc 를 입력할 때까지 수정한다.(3cw: 커서 위치부터 세 글자 수정)|

## 🔹vi 사용법_복사 및 잘라 붙이기

| :---: | :---: | :--- |

|:#d||#로 지정한 행을 잘라냄. 3d는 세 번째 행을 잘라냄|
|:<범위>d||범위로 지정한 행을 잘라냄.1,4d는 1~4행을 잘라냄|
|:pu||현재 행 다음에 버피의 내용을 붙인다|
|:#pu||#로 지정한 행 다음에 버피의 내용을 붙인다. 4pu|
||||

## 🔹vi 사용법_검색 및 바꾸기
| 명령키(명령어)|  | 기능|

||||

|:%s/문자열1/문자열2/g||파일 전체에서 모든 문자열 1을 문자열 2로 바꿈.|

|:<범위>s/문자열1/문자열2/g||범위 내 모든 행에서 문자열 1을 문자열 2로 바꿈|

||||


## 🔹vi 사용법_환경 설정 명령
| 명령키(명령어)|  | 기능|
| :---: | :---: | :--- |
|set nu||파일 내용의 각 행에 행 번호를 표시한다.(보이기만 할 뿐 저장되지는 않는다.)|
|set nonu||행 번호를 감춘다.|
|set list||눈에 보이지 않는 특수문자를 표시함|
|set showmode||현재 모드를 표시|
|set noshowmode||현재 모드를 감춤|
|set||set으로 설정한 모든 vi 환경 설정 값을 출력함|
|set all||모든 vi 환경 변수와 현재 값을 출력함|
