````python
# 👨‍🏫 for구문을 while 구문으로 변경해서 해보기
print("{}의 약수 : 1, ".format(num), end='')

i = 2
while i < num:  # i != num
    #(i가 num과 같지 않을 때까지, 연산 속도 차이가 많이 남-> 자주 쓰는 기호 not always)
    # pass    # 코드 작서 못하고 있으니 우선 pass 작성할 수 있다
    if num % i == 0:
        print("{}, ".format(i), end = '')
    i = i + 1

print(num)
````

````python
# 👨‍🏫 break 활용
print("{}의 약수 : 1, ".format(num), end='')

i = 2
while True:
    if num % i == 0:
        print("{}, ".format(i), end = '')

    i = i + 1

    if i == num:
        break

print(num)
````

````python
# 👨‍🏫 == 를 != 로 변경
print("{}의 약수 : 1, ".format(num), end='')
i = 2
while i < num:  # i != num
    if num % i != 0:
            i = i + 1            # i = i + 1 을 적지 않으면 아래 'print, i = i+1'이 skip 된다
            continue

    print("{}, ".format(i), end = '')
    i = i + 1

print(num)
````

````python
def absolute_value(num):                                 # 함수 선언
    """This function returns the absolute
    value of the entered number"""
    if num >= 0:
        return num
    else:
        return -num
num = absolute_value(2)
print(num)                                        # 함수에 2 대입한 값 출력
print(absolute_value(-4))                         # 함수에 -4 대입한 값 출력
````

📌 Default arguments
- 함수 선언 시 default 지정해줌
- 매개변수에 default를 지정했다면, 그 이후에 오는 모든 인자들에는 default 값이 있어야 한다.

````python
def greet(name, msg="Good morning!"):         # msg값은 "Good morning!"으로 default 값 지정해줌
    """
    This function greets to
    the person with the
    provided message.
    If the message is not provided,
    it defaults to "Good
    morning!"
    """
    print("Hello", name + ', ' + msg)
greet("Kate")                                   # name에 "Kate"를 넣고, msg는 default 값 출력
greet("Bruce", "How do you do?")                # msg에 default 값은 있지만 name,msg에 각각의 값을 넣어 출력
````

````python
def greet(msg = "Good morning!", name):       # msg에 default 값을 선언함. default 지정한 이후에는 모두 default 값을 선언해야 함
    print("Hello", name + ', ' + msg)
    # default 선언 후 지정하지 않은 매개 변수값이 있어 문법상 오류
greet("Kate")
greet("Bruce", "How do you do?")
````

📌 Keyword arguments

````python
def greet(name, msg="Good morning!"):                  # 함수 선언, msg는 default값 설정
    """                                                # keyword를 name, msg로 설정함
    This function greets to
    the person with the
    provided message.
    If the message is not provided,
    it defaults to "Good
    morning!"
    """
    print("Hello", name + ', ' + msg)

greet(name = "Bruce",msg = "How do you do?")     # keyword 대로 name에 Bruce, msg에 How do you do 입력
greet(msg = "How do you do?",name = "Bruce")     # msg, name의 순서는 다르게 입력되어 있지만 keyword에 맞게 입력
greet("Bruce", msg = "How do you do?")           # keyword 설정 안된 Bruce를 첫번째에 넣어줌, 뒤에 keyword에 맞게 입력
greet(name = "Bruce", "How do you do?")          # name= keyword 설정됨,이후 선언하지 않은 keyword가 있어서 문법적 오류 발생
````

📌 Arbitrary arguments
- 인자에 대한 임의의 개수를 반영가능(tuple 변환 발생)(cause tuple은 순서 변경이 불가하기 때문에!)
별표(asterisk)*를 인자명 앞에 사용

````python
def greet(*names):                        # '*'을 적어 '몇 개를 받을 지 모르지만 다 넣어줄게'로 해석
                                          # 자료구조형 이라 생각할 수 있음
                                          # (tuple 변환 일어남 <- tuple은 순서가 고정되어 있기 때문에 변환함)
    """This function greets all
    the person in the names tuple."""

    # names is a tuple with arguments
    for name in names:                   # names 를 name라는 변수에 담아 names에 더이상 없을 때까지 반복해라
        print("Hello", name)
greet("Monica", "Luke", "Steve", "John")
````

📌 Recursion(재귀함수)
- 자기 자신을 호출하는 것
- 최대 재귀 깊이(depth)는 1000으로 설정되어 있다
- (특별한 경우 아니면 무한루프 빠질 수 있으니 다른 함수 사용하는 것을 추천함)

````python
def factorial(x):
    """This is a recursive function
    to find the factorial of an integer"""

    if x == 1:
      return 1
    else:
      return (x * factorial(x-1))

num = 3
print("The factorial of", num, "is", factorial(num))
````

````python
x = "global"
def foo():
    x = x * 2          # x값이 새롭게 x+2로 새롭게 정의됨
    print(x)           # global+2가 정의 안됨-> x 값 오류
foo()
# Error
````

````python
x = "global "
def foo():
    global x
    y = "local"
    x = x * 2
    print(x)
    print(y)
foo()
````

📌 Nonlocal variable
- Local 범위가 정의되지 않은 nested 함수에서 사용
- Local 또는 global 범위에 있을 수 없음

````python
def foo():
    y = "local"
    print(y)
foo()
````

````python
def foo():
    y = "local"
foo()
print(y)
# Error
````

````python
def outer():
    x = "local"
    def inner():
        nonlocal x
        x = "nonlocal"
        print("inner:", x)
    inner()
    print("outer:", x)
outer()
````
